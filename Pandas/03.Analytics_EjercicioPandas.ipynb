### Estudia donde faltan datos, y propón una estrategia de reemplazo o gestión. Puedes usar la información que has sacado de los propios datos

```python
dfm = pd.read_excel(os.getcwd() + "/SalariesExtended_modified.xlsx")
dfm

# Identificar las columnas con datos faltantes
missing_data = dfm.isnull().sum()

# Mostrar las columnas con datos faltantes y el número de valores faltantes en cada una
print("Columnas con datos faltantes y el número de valores faltantes en cada una:")
print(missing_data[missing_data > 0])

# Estrategia de reemplazo o gestión
# Para las columnas numéricas, podemos reemplazar los valores faltantes con la media de la columna
# Para las columnas categóricas, podemos reemplazar los valores faltantes con la moda de la columna

# Reemplazar valores faltantes en columnas numéricas con la media
numeric_columns = dfm.select_dtypes(include=['number']).columns
dfm[numeric_columns] = round(dfm[numeric_columns].fillna(dfm[numeric_columns].mean()),2)

# Reemplazar valores faltantes en columnas categóricas con la moda
categorical_columns = dfm.select_dtypes(include=['object']).columns
dfm[categorical_columns] = dfm[categorical_columns].apply(lambda x: x.fillna(x.mode()[0]))

# Guardar el DataFrame modificado en un nuevo archivo Excel
dfm.to_excel(os.getcwd() + '/SalariesExtended_filled.xlsx', index=False)

print("Los valores faltantes han sido gestionados y el archivo modificado se ha guardado como 'SalariesExtended_filled.xlsx'.")
```

#### Crea tablas resumen que crucen las variables categóricas del dataset

```python
# Salarios por Sexo y Provincia

# Cargar el archivo Excel
dfr = pd.read_excel(os.getcwd() + '/SalariesExtended_with_preg_mean.xlsx')


# Identificar las columnas categóricas
categorical_columns = dfr.select_dtypes(include=['object']).columns

# Crear tablas resumen que crucen las variables categóricas
summary_tables = {}
for col1 in categorical_columns:
    for col2 in categorical_columns:
        if col1 != col2:
            summary_table = pd.crosstab(dfr[col1], dfr[col2])
            summary_tables[f'{col1}_vs_{col2}'] = summary_table

# Mostrar las tablas resumen
for key, table in summary_tables.items():
    print(f"Tabla resumen: {key}")
    print(table)
    print("\n")
```

### Gestión de fechas

Homogeniza el formato de las fechas de inicio y fin. Puedes hacerlo de muchas maneras distintas. El paquete datetime puede ser de gran ayuda ( https://stackabuse.com/converting-strings-to-datetime-in-python/ o https://stackoverflow.com/questions/466345/converting-string-into-datetime ).

```python
from datetime import datetime

# Cargar el archivo Excel
dffecha = pd.read_excel(os.getcwd() + '/SalariesExtended_filled.xlsx')

# Función para homogenizar el formato de la fecha
def homogenize_date_format(date):
    if isinstance(date, int):
        # Convertir enteros a cadenas
        date_str = str(date)
    else:
        date_str = date
    
    # Definir posibles formatos de fecha
    date_formats = [
        "%Y-%m-%d",
        "%d/%m/%Y",
        "%m-%d-%Y",
        "%d-%m-%Y",
        "%Y/%m/%d"
    ]
    
    for fmt in date_formats:
        try:
            # Intentar analizar la cadena de fecha con el formato actual
            date_obj = datetime.strptime(date_str, fmt)
            # Devolver la fecha en el formato deseado
            return date_obj.strftime("%Y-%m-%d")
        except ValueError:
            continue
    
    # Si ningún formato coincide, devolver la cadena original
    return date_str

# Combinar startday, startmonth y startyear en una sola columna start_date
dffecha['start_date'] = dffecha.apply(lambda row: f"{row['startyear']}-{row['startmonth']:02d}-{row['startday']:02d}", axis=1)

# Aplicar la función a las columnas start_date y end
dffecha['start_date'] = dffecha['start_date'].apply(homogenize_date_format)
dffecha['end'] = dffecha['end'].apply(lambda x: homogenize_date_format(str(x)))

# Guardar el DataFrame modificado de nuevo en un archivo Excel
output_file_path = 'SalariesExtended_filled_homogenized.xlsx'
dffecha.to_excel(output_file_path, index=False)

print(f"Las fechas han sido homogenizadas y guardadas en {output_file_path}.")
```

#### Crea una columna que sea el tiempo total de contrato, usando los datos de las columnas ya existentes

```python
# Cargar el archivo Excel
dffecha2 = pd.read_excel(os.getcwd() + '/SalariesExtended_filled_homogenized.xlsx')

# Convertir las columnas de fecha a formato datetime
dffecha2['start_date'] = pd.to_datetime(dffecha2['start_date'])
dffecha2['end'] = pd.to_datetime(dffecha2['end'])

# Calcular el tiempo total de contrato en días y añadirlo como una nueva columna
dffecha2['total_contract_time'] = (dffecha2['end'] - dffecha2['start_date']).dt.days

# Guardar el DataFrame modificado de nuevo en un archivo Excel
output_file_path = 'SalariesExtended_contract_time.xlsx'
dffecha2.to_excel(output_file_path, index=False)

print(f"Las fechas han sido homogenizadas y el tiempo total de contrato ha sido calculado y guardado en {output_file_path}.")
```

### Filtrar y exportar

#### Creamos un dataset fitrando solo datos de Barcelona, otro para Girona y otro para resto y exportamos cada uno en una hoja de un libro excel

```python
# Cargar el archivo Excel
dffecha3 = pd.read_excel(os.getcwd() + '/SalariesExtended_contract_time.xlsx')


# Filtrar datos para Barcelona, Girona y el resto
df_barcelona = dffecha3[dffecha3['prov'] == 'Barcelona']
df_girona = dffecha3[dffecha3['prov'] == 'Girona']
df_resto = dffecha3[~dffecha3['prov'].isin(['Barcelona', 'Girona'])]

# Crear un nuevo objeto Excel writer
output_file_path = 'Filtered_Salaries.xlsx'
with pd.ExcelWriter(output_file_path, engine='openpyxl') as writer:
    # Escribir cada DataFrame en una hoja diferente
    df_barcelona.to_excel(writer, sheet_name='Barcelona', index=False)
    df_girona.to_excel(writer, sheet_name='Girona', index=False)
    df_resto.to_excel(writer, sheet_name='Resto', index=False)

print(f"Los datasets filtrados han sido guardados en {output_file_path}.")
```
