Carga el dataset de vuelos "flights14.csv"

```python
import pandas as pd

flights = pd.read_csv("./datos/flights14.csv", index_col=0)
flights.head()
```

1. escribe un programa Pandas para dividir los datos basados en origen u destino. Muestra cada grupo y el número de elementos por grupo

```python
df = pd.DataFrame(flights)

# Dividir los datos basados en 'origin'
grupos_origen = df.groupby('origin')
print("Grupos basados en 'origin':")
for nombre_origen, grupo_origen in grupos_origen:
    print(f"\nGrupo: {nombre_origen}")
    print(grupo_origen)
    print(f"Número de elementos en el grupo: {len(grupo_origen)}")

# Dividir los datos basados en 'dest'
grupos_destino = df.groupby('dest')
print("\nGrupos basados en 'destino':")
for nombre_destino, grupo_destino in grupos_destino:
    print(f"\nGrupo: {nombre_destino}")
    print(grupo_destino)
    print(f"Número de elementos en el grupo: {len(grupo_destino)}")
```

2. Escribe un programa Pandas para calcula la media, desviación estándar, min y max del retraso total de cada orgige-destino.

```python
df = pd.DataFrame(flights)

# Calcular el retraso total
df['total_delay'] = df['dep_delay'] + df['arr_delay']

# Agrupar por origen y destino y calcular las estadísticas
grouped = df.groupby(['origin', 'dest'])['total_delay'].agg(['mean', 'std', 'min', 'max'])

print(grouped)
```

3. Escribe un programa Pandas para dividir un conjunto de datos y agruparlos por dos columnas, y luego ordenar los resultados agregados dentro de los grupos.

En el siguiente conjunto de datos de vuelos, agrupar "date" (fecha completa) y "carrier". Luego ordenar la suma de 'total_delay' dentro de los grupos.

```python
df = pd.DataFrame(flights)

# Crear una columna de fecha completa
df['date'] = pd.to_datetime(df[['year', 'month', 'day']])

# Calcular el retraso total
df['total_delay'] = df['dep_delay'] + df['arr_delay']

# Agrupar por fecha y carrier y calcular la suma de total_delay
grouped = df.groupby(['date', 'carrier'])['total_delay'].sum().reset_index()

# Ordenar los resultados agregados dentro de los grupos
sorted_grouped = grouped.sort_values(by=['date', 'total_delay'], ascending=[True, False])

print(sorted_grouped)
```

4. Escribe un programa Pandas para contar el múmero de retrasos de salida por mes y origen.

```python
df = pd.DataFrame(flights)

# Filtrar los vuelos con retraso de salida
delayed_flights = df[df['dep_delay'] > 0]

# Contar el número de retrasos de salida por mes y origen
delay_counts = delayed_flights.groupby(['month', 'origin']).size().reset_index(name='delay_count')

print(delay_counts)
```

5. Escribe un programa Pandas para cargar los datos de productos y ventas del archivo FoodMarket.xlsx y mézclalos en un solo dataframe.

```python
import pandas as pd


# Cargar los datos de productos y ventas del archivo FoodMarket.xlsx
file_path = './datos/FoodMarket.xlsx'

# Leer las hojas del archivo Excel
productos_df = pd.read_excel(file_path, sheet_name='Products', engine='openpyxl')
ventas_df = pd.read_excel(file_path, sheet_name='Sellers', engine='openpyxl')

# Mezclar los datos en un solo DataFrame
merged_df = pd.concat([ventas_df, productos_df], axis=1)

merged_df.head()
```

6. Escribe un programa Pandas para extraer información sobre el número de compras totales y el monto total por nombre de producto y año. 

```python
import pandas as pd

# Cargar los datos del archivo FoodMarket.xlsx
file_path = './datos/FoodMarket.xlsx'

# Leer las hojas 'Purchases' y 'Products' del archivo Excel
purchases_df = pd.read_excel(file_path, sheet_name='Purchases', engine='openpyxl')
products_df = pd.read_excel(file_path, sheet_name='Products', engine='openpyxl')

# Convertir la columna 'Fecha' a tipo datetime
purchases_df['Fecha'] = pd.to_datetime(purchases_df['Fecha'])

# Extraer el año de la columna 'Fecha'
purchases_df['Año'] = purchases_df['Fecha'].dt.year

# Unir los DataFrames en función de la columna 'Product'
merged_df = pd.merge(purchases_df, products_df, on='Product', how='inner')

# Calcular el monto total por nombre de producto y año
merged_df['Monto Total'] = merged_df['Quantity'] * merged_df['Price']

# Agrupar por nombre de producto y año y calcular el número de compras totales y el monto total
result = merged_df.groupby(['Name.x', 'Año']).agg({'Quantity': 'sum', 'Monto Total': 'sum'}).reset_index()

result
```
